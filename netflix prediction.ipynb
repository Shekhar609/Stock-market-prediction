# Outlier Detection
#Outlier Detection for k-mean null value replaced dataset

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select the columns you want to use for outlier detection (X)
X = data_withoutMD[['Open', 'High', 'Low']]

# Standardize the data to bring all features to the same scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled 
# Create a PCA object
pca = PCA()

# Fit the PCA model and transform the data
X_pca = pca.fit_transform(X_scaled)
# Plot the explained variance ratio to decide on the number of components
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

plt.plot(cumulative_variance)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
print((X_pca))

dbscan = DBSCAN(eps=0.5, min_samples=5)

dbscan.fit(X_scaled)

# Get the cluster labels (-1 indicates noise/outliers)
labels = dbscan.labels_

# Create a new column in the DataFrame to store the outlier labels
data_withoutMD['outlier'] = (labels == -1)

# The 'outlier' column now contains True for outliers and False for inliers.

outliers = data_withoutMD[data_withoutMD['outlier'] == True]

# Print the outliers
print(outliers)
```

```python
# # Filter and print the rows that are not outliers (outlier=False)
outlier_free_data = data_withoutMD[data_withoutMD['outlier'] == False]

# # Print the outlier-free data
outlier_free_data.info()
```

```python
#Choose the number of components (e.g., n_components=2)
pca = PCA(n_components=2)
X_final = pca.fit_transform(X_scaled)
# Access the principal components
principal_components = pca.components_
plt.scatter(X_final[:, 0], X_final[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
